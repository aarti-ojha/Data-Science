```python
import pandas as pd
```

```python
df = pd.read_csv('Pokemon.csv')
```

```python
new_df= df.loc[(df['Type 1']=='Grass' )| (df['Type 2']=='Poison') & (df['HP']>70)]
#new_df.to_csv('filtered.csv')
print(new_df)
```

```python
df.loc[df['Name'].str.contains('Mega')]
```

```python
#To see or check two values for one field we need to install regex function

import re
df.loc[df['Type 1'].str.contains('fire|grass',flags=re.I , regex=True)]

#flags ignore the capital and small keys
```

```python
#Get the names that start with PI/Pi/pi

df.loc[df['Name'].str.contains('^pi[a-z]*',flags=re.I , regex=True)]
#* means zero or more charecters

```

# Conditional changes

```python
df.loc[df['Type 1']=='Fire','Type 1']='Water'
df
```

```python
df.loc[df['Total']>500,['Generation','Legendary']]='Super'
df
```
